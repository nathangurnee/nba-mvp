```python
## team win-loss ##
team_12 = pd.read_csv('../data/team/12_13_team.csv')
team_12.head()
```

```python
team_12 = team_12[['Team', 'W/L%']]
team_12.head()
```

```python
# team abbreviations
teams = {
    'Atlanta Hawks': 'ATL',
    'Brooklyn Nets': 'BRK',
    'Boston Celtics': 'BOS',
    'Charlotte Hornets': 'CHO',
    'Charlotte Bobcats': 'CHA',
    'Chicago Bulls': 'CHI',
    'Cleveland Cavaliers': 'CLE',
    'Dallas Mavericks': 'DAL',
    'Denver Nuggets': 'DEN',
    'Detroit Pistons': 'DET',
    'Golden State Warriors': 'GSW',
    'Houston Rockets': 'HOU',
    'Indiana Pacers': 'IND',
    'Los Angeles Clippers': 'LAC',
    'Los Angeles Lakers': 'LAL',
    'Memphis Grizzlies': 'MEM',
    'Miami Heat': 'MIA',
    'Milwaukee Bucks': 'MIL',
    'Minnesota Timberwolves': 'MIN',
    'New Orleans Pelicans': 'NOP',
    'New Orleans Hornets': 'NOH',
    'New York Knicks': 'NYK',
    'Oklahoma City Thunder': 'OKC',
    'Orlando Magic': 'ORL',
    'Philadelphia 76ers': 'PHI',
    'Phoenix Suns': 'PHO',
    'Portland Trail Blazers': 'POR',
    'Sacramento Kings': 'SAC',
    'San Antonio Spurs': 'SAS',
    'Toronto Raptors': 'TOR',
    'Utah Jazz': 'UTA',
    'Washington Wizards': 'WAS'
}
```

```python
team_12 = team_12.replace({'Team': teams})
team_12.head()
```

```python
team_12 = team_12.rename(columns = {'Team': 'Tm'})
team_12.head()
```

```python
## player per-game ##
game_12 = pd.read_csv('../data/per_game/12_13_game.csv')
game_12.head()
```

```python
game_12['Player'] = game_12['Player'].str.replace('*', '', regex=False)
game_12.iloc[7]['Player']
```

```python
game_12 = game_12.drop('Player-additional', axis=1)
game_12.head()
```

```python
## player totals ##
total_12 = pd.read_csv('../data/totals/12_13_totals.csv')
total_12.head()
```

```python
total_12['Player'] = total_12['Player'].str.replace('*', '')
total_12.iloc[7]['Player']
```

```python
total_12 = total_12.drop('Player-additional', axis=1)
total_12.head()
```

```python
total_12 = total_12.rename(columns = {col: 'T_' + col for col in total_12.columns if col not in ['Rk', 'Player', 'Pos', 'Age', 'Tm', 'G', 'GS']})
total_12.head()
```

```python
total_12 = total_12.drop(['Rk', 'Pos', 'Age', 'Tm', 'G', 'GS'], axis=1)
total_12.head()
```

```python
season_12 = game_12.merge(total_12, on='Player')
season_12.head()
```

```python
season_12 = season_12.drop('Rk', axis=1)
season_12.head()
```

```python
season_12 = season_12.merge(team_12, on='Tm')
print(season_12.shape)
season_12.head()
```

```python
# lowest points per game -- 13.8
season_12 = season_12[season_12['PTS'] > 12]
print(season_12.shape)
season_12.head()
```

```python
# lowest recorded number of games started by an mvp -- 49 (statmuse)
# will let cutoff be 40
season_12 = season_12[season_12['GS'] > 40]
print(season_12.shape)
season_12.head()
```

```python
# worst win percentage -- 45.8%
# W/L% = W / (W + L)  (basketball-reference.com)
season_12 = season_12[season_12['W/L%'] >= 0.45]
print(season_12.shape)
season_12.head()
```

```python
mvp_12.head()
```

```python
season_12 = season_12.merge(mvp_12, how='left')
print(season_12.shape)
season_12.head()
```

```python
season_12['Share'] = season_12['Share'].fillna(0)
season_12.head()
```

```python
season_12 = season_12.drop(['Pos', 'Tm'], axis=1)
season_12.head()
```

```python
season_12['Season'] = 12
season_12.head()
```
